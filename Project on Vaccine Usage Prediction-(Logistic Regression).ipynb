```python
# Checking for Outliers via visualization:
for col in df.columns:
    if df[col].dtype in ["int64","float64"]:
        plt.figure()
        df.boxplot([col])
        plt.show()
```

```python
# Dealing with Outliers:
def cap_data(df):
    for col in df.columns:
        print("capping the ",col)
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            percentiles = df[col].quantile([0.01,0.99]).values
            df[col][df[col] <= percentiles[0]] = percentiles[0]
            df[col][df[col] >= percentiles[1]] = percentiles[1]
        else:
            df[col]=df[col]
    return df

df=cap_data(df)
```

```python
# Checking for unique values in the desired feature/column
df['h1n1_vaccine'].unique()
```

```python
# Label Encoding on Catagorical columns:
df_encoded=df.copy()
for col in df_encoded.columns:
    if df_encoded[col].dtype=='object':
        df_encoded[col]=df_encoded[col].astype('category')
        df_encoded[col]=df_encoded[col].cat.codes
df_encoded.head()
# Encoding done.
```

### 

### 4. Data Splitting:

```python
# Data splitting
x=df_encoded.drop(labels=['h1n1_vaccine'],axis=1)
y=df_encoded[['h1n1_vaccine']]
```

### 5. Logistic Regression Model Build:

```python
lgr=LogisticRegression()
```

### #6.1 Without Scaling:

```python
# Train & Test Spliting
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# Model Fitting
lgr.fit(x_train, y_train)
```

```python
# Accuracy Testing
lgrtr=lgr.score(x_train,y_train)
print('Train Score:',lgrtr)
lgrte=lgr.score(x_test,y_test)
print('Test Score: ',lgrte)
```

## We have got a decent score for both training and testing data without scaling, we have to check with scaled technique.

### #6.2 With Standard Scaling:

```python
# Standard Scaling
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
# Train & Test Spliting
x_train,x_test,y_train,y_test=train_test_split(x_scaled, y, test_size=0.25, random_state=0)
```

```python
# Model Fitting
lgr.fit(x_train, y_train)
```

```python
# Accuracy Testing
lgrtrs=lgr.score(x_train,y_train)
print('Train Score:',lgrtrs)
lgrtes=lgr.score(x_test,y_test)
print('Test Score: ',lgrtes)
```

## We have got a decent similar score for both training and testing data with and without scaling, we have to check with other scaled technique.

### #6.3 With Min Max Scaling:

```python
# Min Max Scaling
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
# Train & Test Spliting
x_train,x_test,y_train,y_test=train_test_split(x_scaled, y, test_size=0.25, random_state=0)
```

```python
# Model Fitting
lgr.fit(x_train, y_train)
```

```python
# Accuracy Testing
lgrtrm=lgr.score(x_train,y_train)
print('Train Score:',lgrtrm)
lgrtem=lgr.score(x_test,y_test)
print('Test Score: ',lgrtem)
```

## We have got a decent similar score for both training and testing data with and without scaling, we have to check with other scaled technique.

### #6.4 With Robust Scaling:

```python
# Robust Scaling
scaler=RobustScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
# Train & Test Spliting
x_train,x_test,y_train,y_test=train_test_split(x_scaled, y, test_size=0.25, random_state=0)
```

```python
# Model Fitting
lgr.fit(x_train, y_train)
```

```python
# Accuracy Testing
lgrtrr=lgr.score(x_train,y_train)
print('Train Score:',lgrtrr)
lgrter=lgr.score(x_test,y_test)
print('Test Score: ',lgrter)
```

## We have got a decent similar score for both training and testing data with and without scaling, we will see the difference between them in the conclusion segment.
